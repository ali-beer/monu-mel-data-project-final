# Australian Bureau of Statistics
- 2016 Census data on labour force status of one parent and couple parent families with dependent children aged 0-4 years
- 2016 SEIFA Index of Relative Socio-Economic Disadvantage data 

```python
# Modules
import pandas as pd
```

```python
# Import the data into pandas dataframe
one_parent_df = pd.read_csv("2016Census_G45A_VIC_SA2.csv")
one_parent_part2_df = pd.read_csv("2016Census_G45B_VIC_SA2.csv")
couple_df = pd.read_csv("2016Census_G44A_VIC_SA2.csv")
seifa_df = pd.read_csv("SEIFA 2016.csv")
```

```python
# List columns in one parent df
list(one_parent_df.columns)
```

```python
# Rename columns
renamed_one_parent_df = one_parent_df.rename(columns = {'M_LPLFS_EFT_DC_0_4': "Male_Lone_Parent_Employed_Worked_Full_Time",
                                             'M_LPLFS_EPT_DC_0_4': "Male_Lone_Parent_Employed_Worked_Part_Time",
                                             'M_LPLFS_EAW_DC_0_4': "Male_Lone_Parent_Employed_Away_From_Work",
                                             'M_LPLFS_HNS_DC_0_4': "Male_Lone_Parent_Hours_Worked_Not_Stated",
                                             'M_LPLFS_EW_T_DC_0_4': "Male_Lone_Parent_Total_Employed",
                                             'M_LPLFS_ULFTW_DC_0_4': "Male_Lone_Parent_Unemployed_Looking_For_Full_Time_Work",
                                             'M_LPLFS_ULPTW_DC_0_4': "Male_Lone_Parent_Unemployed_Looking_For_Part_Time_Work",
                                             'M_LPLFS_U_T_DC_0_4': "Male_Lone_Parent_Total_Unemployed",
                                             'M_LPLFS_T_LF_DC_0_4': "Male_Lone_Parent_Total_Labour_Force",
                                             'M_LPLFS_NLF_DC_0_4': "Male_Lone_Parent_Not_In_The_Labour_Force",
                                             'M_LPLFS_LFSNS_DC_0_4': "Male_Lone_Parent_Labour_Force_Status_Not_Stated",
                                             'M_LPLFS_T_DC_0_4': "Male_Lone_Parent_Total",
                                             'F_LPLFS_EFT_DC_0_4': "Female_Lone_Parent_Employed_Worked_Full_Time",
                                             'F_LPLFS_EPT_DC_0_4': "Female_Lone_Parent_Employed_Worked_Part_Time",
                                             'F_LPLFS_EAW_DC_0_4': "Female_Lone_Parent_Employed_Away_From_Work",
                                             'F_LPLFS_HNS_DC_0_4': "Female_Lone_Parent_Hours_Worked_Not_Stated",
                                             'F_LPLFS_EW_T_DC_0_4': "Female_Lone_Parent_Total_Employed",
                                             'F_LPLFS_ULFTW_DC_0_4': "Female_Lone_Parent_Unemployed_Looking_For_Full_Time_Work",
                                             'F_LPLFS_ULPTW_DC_0_4': "Female_Lone_Parent_Unemployed_Looking_For_Part_Time_Work",
                                             'F_LPLFS_U_T_DC_0_4': "Female_Lone_Parent_Total_Unemployed",
                                             'F_LPLFS_T_LF_DC_0_4': "Female_Lone_Parent_Total_Labour_Force",
                                             'F_LPLFS_NLF_DC_0_4': "Female_Lone_Parent_Not_In_The_Labour_Force",
                                             'F_LPLFS_LFSNS_DC_0_4': "Female_Lone_Parent_Labour_Force_Status_Not_Stated",
                                             'F_LPLFS_T_DC_0_4': "Female_Lone_Parent_Total",
                                             'T_LPLFS_EFT_DC_0_4': "Total_Lone_Parents_Employed_Worked_Full_Time"})
```

```python
# Preview df
renamed_one_parent_df.head()
```

```python
one_parent_part2_df.head()
```

```python
# List columns
list(one_parent_part2_df.columns)
```

```python
# Rename columns
renamed_one_parent_part2_df = one_parent_part2_df.rename(columns = {'T_LPLFS_EPT_DC_0_4':"Total_Lone_Parents_Employed_Worked_Part_Time",
                                                         'T_LPLFS_EAW_DC_0_4': "Total_Lone_Parents_Employed_Away_From_Work",
                                                         'T_LPLFS_HNS_DC_0_4': "Total_Lone_Parents_Hours_Worked_Not_Stated",
                                                         'T_LPLFS_EW_Tot_DC_0_4': "Total_Lone_Parents_Total_Employed",
                                                         'T_LPLFS_ULFTW_DC_0_4': "Total_Lone_Parents_Unemployed_Looking_For_Full_Time_Work",
                                                         'T_LPLFS_ULPTW_DC_0_4': "Total_Lone_Parents_Unemployed_Looking_For_Part_Time_Work",
                                                         'T_LPLFS_U_T_DC_0_4': "Total_Lone_Parents_Total_Unemployed",
                                                         'T_LPLFS_T_LF_DC_0_4': "Total_Lone_Parents_Total_Labour_Force",
                                                         'T_LPLFS_NLF_DC_0_4': "Total_Lone_Parents_Not_In_The_Labour_Force",
                                                         'T_LPLFS_LFSNS_DC_0_4': "Total_Lone_Parents_Labour_Force_Status_Not_Stated",
                                                         'T_LPLFS_Tot_DC_0_4': "Total_Lone_Parents_Total"})
```

```python
renamed_one_parent_part2_df.head()
```

```python
# Merge one parent dataframes, join on sa2 code
one_parent_merge_df = pd.merge(renamed_one_parent_df, renamed_one_parent_part2_df, on="SA2_MAINCODE_2016", how="outer")
one_parent_merge_df.head()
```

```python
couple_df.head()
```

```python
# List columns in couple df
list(couple_df.columns)
```

```python
# Merge couple df with one parent df, join on sa2 code
family_merge_df = pd.merge(couple_df, one_parent_merge_df, on="SA2_MAINCODE_2016", how="outer")
family_merge_df.head()
```

```python
list(seifa_df.columns)
```

```python
# Rename columns in seifa df
seifa_df = seifa_df.rename(columns = {"2016_sa2_code": "SA2_MAINCODE_2016",
                                      "2016_sa2_name ": "SA2_name"})
seifa_df.head()
```

```python
# Merge family df and seifa df, join on sa2 code
seifa_family_merge_df = pd.merge(family_merge_df, seifa_df, on="SA2_MAINCODE_2016", how="left")
seifa_family_merge_df.head()
```

```python
# Write to csv
seifa_family_merge_df.to_csv("clean_2016_abs.csv", index=False, header=True)
```
