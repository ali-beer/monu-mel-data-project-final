```python
import pandas as pd
import numpy as np
import re
```

```python
#read csv
domain = 'Health'
df = pd.read_excel('raw.xlsx',sheet_name = f'SA2 {domain}')
df.head(10)
```

```python
#find Victoria
states_tempt = [(v,i)for i,v in enumerate(df.iloc[:,1])if re.findall('[a-zA-Z]+', str(v))]
states_tempt
```

```python
#select Vic data, not including the line of the total of Vic
row_idx = list(range(665, 1178))
```

```python
#select year 2018 & code& name
col_idx  =[i for i,v in enumerate(df.iloc[4, :]) if v == 2018]
col_idx = [1, 2] + col_idx
```

```python
#rename header
head = [h for h in df.iloc[4] if isinstance(h, str)]
head = ['code', 'name', 'total', 'on_track', 'at_risk', 'vulnerable']
# #rename header for 'one or more' and 'Two or more' sheets
# head = ['code', 'name', 'total',  'vulnerable']
head
```

```python
df = df.iloc[row_idx, col_idx]
```

```python
df.columns = head
```

```python
df.head(20)
```

```python
# dropape rows contain non numeric
df = df[df.iloc[:,2].apply(lambda x: isinstance(x, int))]
df = df[df.iloc[:,3].apply(lambda x: isinstance(x, int))]

#should be commentted for 'one or more' and 'two or more' sheets
df = df[df.iloc[:,4].apply(lambda x: isinstance(x, int))]
df = df[df.iloc[:,5].apply(lambda x: isinstance(x, int))]
```

```python
# calculate percentage
df['vulnerable_percentage'] = df['vulnerable']/df['total']
df['vulnerable_percentage'] = df['vulnerable_percentage'].apply(lambda x: round(x,2))
df.head(10)
```

```python
# filter subregion
region_total = df.iloc[0, 2]
current_total = 0
row_idx2 = []
# print(region_total)
count = 1
while count < df.shape[0]:
    i = count
    sub_num = df.iloc[i, 2]
    
    if current_total + sub_num <= region_total:
        row_idx2.append(i)
        current_total += sub_num
        count += 1
    elif current_total == region_total and i + 1 < df.shape[0]:
        region_total = df.iloc[i, 2]
        current_total = 0
        count += 1
    else:
        break
    
```

```python
df.head(20)
```

```python
len(row_idx2)
```

```python
df = df.iloc[row_idx2]
df = df.reset_index(drop=True)
```

```python
df.to_csv(f'{domain.lower()}.csv', index=False)
```
